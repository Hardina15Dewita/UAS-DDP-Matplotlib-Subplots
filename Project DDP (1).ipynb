```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f'matplotlib: {matplotlib.__version__}')
print(f'numpy: {np.__version__}')
```

```python
# SIMPLE LINE PLOT
```

```python
# Data suhu harian selama seminggu di Ciputat

hari = np.array(['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'])
suhu = np.array([34, 32, 36, 32, 35, 31, 33])
```

```python
# Plotting simple line plot
plt.plot(hari, suhu, marker='o', linestyle='-', color='blue')

# Menambahkan label pada sumbu
plt.xlabel('Hari')
plt.ylabel('Suhu')

# Menambahkan judul plot
plt.title('Perkembangan Suhu Harian Selama Seminggu')

# Menampilkan grid untuk memudahkan pembacaan
plt.grid(True)

# Menampilkan plot
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f'matplotlib: {matplotlib.__version__}')
print(f'numpy: {np.__version__}')
```

```python
# Data Penggunaan listrik dan air per bulan
bulan = np.arange(1,7)
listrik = np.array([150, 140, 160, 130, 170, 150])
air = np.array([10, 12, 15, 11, 14, 13])
```

```python
# Membuat figure dan axes (subplots) menggunakan OO style
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,6))

# Plotting penggunaan listrik
axes[0].plot(bulan, listrik, marker='o', color='blue')
axes[0].set_title('Penggunaan Listrik per Bulan')
axes[0].set_xlabel('Bulan')
axes[0].set_ylabel('Penggunaan Listrik (kWh)')

# Menampilkan grid untuk memudahkan pembacaan
plt.grid(True)

# Plotting penggunaan air
axes[1].plot(bulan, air, marker='o', color='green')
axes[1].set_title('Penggunaan Air per Bulan')
axes[1].set_xlabel('Bulan')
axes[1].set_ylabel('Penggunaan Air (m3)')

# Menampilkan grid untuk memudahkan pembacaan
plt.grid(True)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(f'matplotlib: {matplotlib.__version__}')
print(f'numpy: {np.__version__}')
```

```python
# Data waktu tidur dan waktu berolahraga per hari dalam seminggu 
hari = np.array (['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu'])
waktu_tidur = np.array([7, 6.5, 7, 7.5, 6, 8, 8.5])
waktu_olahraga = np.array([0.5, 1, 1.5, 1, 0.5, 2, 1])
```

```python
# Membuat multiple subplots dengan pyplot style
plt.figure(figsize=(10,6))

# Subplot pertama : Waktu Tidur
plt.subplot(2, 1, 1)
plt.plot(hari, waktu_tidur, marker='o', color='blue')
plt.title('Waktu Tidur per Hari')
plt.xlabel('Hari')
plt.ylabel('Jam')

# Subplot kedua : Waktu Berolahraga
plt.subplot(2, 1, 2)
plt.plot(hari, waktu_olahraga, marker='o', color='green')
plt.title('Waktu Berolahraga per Hari')
plt.xlabel('Hari')
plt.ylabel('Jam')

# Menampilkan grid untuk memudahkan pembacaan
plt.grid(True)

# Menampilkan plot
plt.tight_layout()
plt.show()
```
